## Practice with Folium 

Below is a snippet of code to display the neighborhoods with the python module folium. It takes in the **neighbourhoods.json** file from Moodle. This displays the neighborhoods from the Airbnb dataset. Free free to mess around with the code and change things around.


See http://folium.readthedocs.io/en/latest/quickstart.html for more practice with folium.

```python
import json
import folium 

# Creates a folium map
map_osm = folium.Map(location=[39.7392,-104.9903])

# Loads/opens json file
listings = json.load(open('neighbourhoods.json'))

# Creates a folium geojson and adds it to the map
folium.GeoJson(listings).add_to(map_osm)


# Display map
map_osm
```
